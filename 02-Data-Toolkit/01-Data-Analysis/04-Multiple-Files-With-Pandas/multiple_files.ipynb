## Merge Practice

```python
a_df = pd.DataFrame({
    'Country': ['Germany', 'France', 'Belgium', 'Finland'],
    'Population (M)': [82.8, 67.2, 11.4, 5.5],
    'Capital': ['Berlin', 'Paris', 'Brussels', 'Helsinki']
})
a_df
```

```python
b_df = pd.DataFrame({
    'Country': ['Germany', 'France', 'Belgium', 'Canada'],
    'HDI': [0.936, 0.901, 0.916, 0.926]
})
b_df
```

### Inner Merge Practice

```python
inner_merged_df = a_df.set_index('Country')
inner_merged_df = inner_merged_df.merge(b_df, on='Country')
inner_merged_df
```

#### Check your code

## Olympic Sports and Medals, 1896-2014

```python
countries_df = pd.read_csv('dictionary.csv')
countries_df
```

```python
summer_df = pd.read_csv('summer.csv')
summer_df
```

```python
winter_df = pd.read_csv('winter.csv')
winter_df
```

### Combining The Data

```python
summer_countries_df = countries_df.set_index('Code')
summer_countries_df = summer_countries_df.merge(summer_df.set_index('Country'), left_index=True, right_index=True)
summer_countries_df['Season'] = 'Summer'
summer_countries_df
```

```python
winter_countries_df = countries_df.set_index('Code')
winter_countries_df = winter_countries_df.merge(winter_df.set_index('Country'), left_index=True, right_index=True)
winter_countries_df['Season'] = 'Winter'
winter_countries_df
```

```python
all_df = pd.concat([summer_countries_df, winter_countries_df])
all_df.reset_index(inplace=True)
all_df = all_df.rename(columns={'index':'Code'})
all_df
```

#### Check your code

```python
from nbresult import ChallengeResult

result = ChallengeResult('all_df',
    all_df_shape=all_df.shape,
    all_df_columns=set(all_df.columns)
)
result.write()

print(result.check())
```

### Top Countries Analysis

```python
top_10_df = pd.DataFrame()
top_10_df['Country'] = all_df[ all_df['Year']>=1984 ]['Country']
top_10_df['Medal Count'] = 1
top_10_df = top_10_df.groupby('Country').count().sort_values('Medal Count', ascending=False)[:10]

top_10_df
```

```python
top_10_df.plot(kind="bar")
```

#### Check your code

```python
from nbresult import ChallengeResult

result = ChallengeResult('olympic_games',
    top_country_1=top_10_df.iloc[0]['Medal Count'],
    top_country_10=top_10_df.iloc[9]['Medal Count']
)
result.write()

print(result.check())
```

```python
top_10_season_df = pd.DataFrame()

top_10_season_df['Summer'] = all_df[ all_df['Year']>=1984 ][ all_df['Season']=='Summer' ]['Country'].groupby(all_df['Country']).count()

top_10_season_df['Winter'] = all_df[ all_df['Year']>=1984 ][ all_df['Season']=='Winter' ]['Country'].groupby(all_df['Country']).count()

top_10_season_df['Total'] = top_10_season_df['Summer'] + top_10_season_df['Winter']

top_10_season_df = top_10_season_df.sort_values('Total', ascending=False)[:10]

top_10_season_df
```

```python
top_10_season_df[['Summer', 'Winter']].plot(kind="bar")
```

#### Check your code

```python
from nbresult import ChallengeResult

result = ChallengeResult('olympic_games_season',
    top_country_season_shape=top_10_season_df.shape,
    top_country_1_summer=top_10_season_df.iloc[0]['Summer'],
    top_country_10_winter=top_10_season_df.iloc[9]['Winter']
)
result.write()

print(result.check())
```
