## Yearly Gas production

```python
yearly_gas_df = gas_df
yearly_gas_df = yearly_gas_df.rename(columns={'Month':'Year'})
yearly_gas_df['Year'] = pd.to_datetime(yearly_gas_df['Year'])
yearly_gas_df['Year'] = yearly_gas_df['Year'].dt.year
yearly_gas_df = yearly_gas_df.groupby(yearly_gas_df['Year']).sum()
yearly_gas_df
```

### Check your code

```python
index_year = yearly_gas_df.index[0]
yearly_gas_shape = yearly_gas_df.shape
us_total = yearly_gas_df.iloc[0,0]
```

```python
from nbresult import ChallengeResult

result = ChallengeResult('full_gas',
    index_year=index_year,
    yearly_gas_shape=yearly_gas_shape,
    us_total=us_total
)
result.write()

print(result.check())
```

```python
plot = yearly_gas_df.filter(items=['U.S.']).plot(kind="bar")
plot.set_xlabel("Year")
```

## Comparing with Crude Oil Production

```python
file = "data/U.S._crude_oil_production.csv"
oil_df = pd.read_csv(file, decimal=",")
oil_df.head(10)
```

---

## Yearly Oil production

```python
yearly_oil_df = oil_df
yearly_oil_df['Month'] =  pd.to_datetime(yearly_oil_df['Month'])
yearly_oil_df['Month'] = yearly_oil_df['Month'].dt.year
yearly_oil_df = yearly_oil_df.rename(columns={'Month':'Year'})
yearly_oil_df = yearly_oil_df.groupby(yearly_oil_df['Year']).sum()
yearly_oil_df
```

```python
yearly_oil_df.filter(items=['U.S. Crude Oil ']).plot(kind='bar')
```

```python
filtered_yearly_oil_df = yearly_oil_df[1:-1]
```

### Check your code

```python
from nbresult import ChallengeResult

result = ChallengeResult('oil',
    filtered_oil_shape=filtered_yearly_oil_df.shape,
    filtered_oil_index_year=filtered_yearly_oil_df.index[0],
    us_total=filtered_yearly_oil_df.iloc[0,0]
)
result.write()

print(result.check())
```

## Merging Oil and Gas Production

```python
total_gas = filtered_yearly_gas_df.filter(items=['Year','U.S.'])
total_oil = filtered_yearly_oil_df.filter(items=['Year','U.S. Crude Oil '])
```

```python
merged_df = pd.concat([total_gas,total_oil], axis=1)
merged_df = merged_df.rename(columns={'U.S':'Gas', 'U.S. Crude Oil ':'Crude Oil'})
merged_df
```

```python
plot = merged_df.plot(kind="bar")
plot.set_xlabel("Year")
plot.legend(['Gas (Millions of Cubic feet)', 'Crude Oil (Thousands of barrels)'])
```
